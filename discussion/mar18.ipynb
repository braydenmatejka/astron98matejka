```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fourCurves():
    x = np.linspace(0,3,100)
    yerr = 0.05 + 2*x
    
    plt.figure()
    i = 2
    while i < 6:
        y = x**i
        plt.errorbar(x,y,yerr=yerr,label=("x^"+str(i)))
        i+=1
    plt.title("Exponential Plots")
    plt.xlabel("x-axis")
    plt.ylabel("Exponents")
    plt.legend()
```

```python
fourCurves()
```

```python
def eSubplots():
    x = np.linspace(0,5,100)
    y = [np.exp(x), np.exp(2*x), np.exp(x)+1, np.exp(x)-1]
    fig,axs = plt.subplots(2,2)
    axs[0,0].plot(x,y[0])
    axs[0,1].plot(x,y[1])
    axs[1,0].plot(x,y[2])
    axs[1,1].plot(x,y[3])
    plt.tight_layout()
    plt.show()
```

```python
eSubplots()
```
