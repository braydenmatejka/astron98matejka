```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
P, V = np.meshgrid(np.linspace(1,10,100), np.linspace(10,30,100))
R = 0.083144
a = 16.02
b = 0.1124
T = ((P + (a/V**2))*(V - b))/R

plt.figure()
plt.imshow(T, cmap='inferno', extent=[1,10,10,30])
plt.xlabel('Pressure')
plt.ylabel('Volume')
plt.colorbar(label='Temperature')
plt.show()
```
