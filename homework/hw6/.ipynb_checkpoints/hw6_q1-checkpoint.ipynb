```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
rand1 = np.random.randint(0,100,40)
rand2 = np.random.randint(0,100,40)
rand3 = np.random.randint(0,100,40)
```

```python
x = np.linspace(1,40,40)
fig, axs = plt.subplots(2,1)
axs[0].plot(x, rand1, c='orange', linewidth=10)
axs[0].plot(x, rand2, 'r--')
axs[1].scatter(x, rand3, c='magenta', marker='*')
```
