```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
N = int(1e3)

x, y = np.random.uniform(0,1,N), np.random.uniform(0,1,N)

xin, yin = [], []
count = 0
for i in range(0,N):
    distance = (x[i]**2 + y[i]**2)**(1/2)
    if distance <= 1:
        count += 1
        xin.append(x[i])
        yin.append(y[i])

fig, axs = plt.subplots(4,1, figsize=(10,40))

for i in range(0,4):
    N = int(10**(i+3))
    x, y = np.random.uniform(0,1,N), np.random.uniform(0,1,N)
    xin, yin = [], []
    
    count = 0
    for j in range(0,N):
        distance = (x[j]**2 + y[j]**2)**(1/2)
        if distance <= 1:
            count += 1
            xin.append(x[j])
            yin.append(y[j])
    
    axs[i].scatter(x, y, s=0.1)
    axs[i].scatter(xin, yin, s=0.1, c='r')
    axs[i].set_title('π ≈ '+str(4*count/10**(i+3)))
    
plt.tight_layout()
plt.show()
    
```
